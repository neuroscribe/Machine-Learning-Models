#### Removing Outliers

```python
data=df["age"]
data=df["gender"]
data=df["impluse"]
data=df["pressurehight"]
data=df["pressurelow"]
data=df["glucose"]
for lk6 in range(len(data)):
    if data[lk6]>276.75:
        df["glucose"].replace(data[lk6],276.75,inplace=True)
    else:
        pass

data=df["kcm"]
for lk6 in range(len(data)):
    if data[lk6]>12.030000000000001:
        df["kcm"].replace(data[lk6],12.030000000000001,inplace=True)
    else:
        pass

data=df["troponin"]
for lk6 in range(len(data)):
    if data[lk6]>0.20475:
        df["troponin"].replace(data[lk6],0.20475,inplace=True)
    else:
        pass

row_drp_rmv_outl=[896, 1156, 1285, 645, 7, 1035, 12, 781, 21, 1302, 411, 796, 1181, 30, 287, 1308, 667, 930, 670, 673, 926, 682, 427, 428, 1069, 429, 1202, 311, 186, 1213, 63, 1219, 836, 1096, 457, 842, 459, 969, 717, 1228, 848, 89, 346, 857, 220, 221, 222, 223, 351, 996, 486, 999, 744, 1003, 1267, 1012, 1140, 125]
df=df.drop(row_drp_rmv_outl).reset_index(drop=True)
```

```python
from sklearn.preprocessing import StandardScaler
target_col_name='class'
col_for_scaled=[]
all_col_nm=X.select_dtypes(include=['int64','int32','float64','float32']).columns.tolist()

for i in range(len(all_col_nm)):
    xx=list(X[all_col_nm[i]].unique())

    if max(xx)>2:
        if all_col_nm[i]==target_col_name:
            pass
        else:
            col_for_scaled.append(all_col_nm[i])
    else:
        pass

x_cols_for_scale=X[col_for_scaled]
scaler = StandardScaler()
scaler.fit(x_cols_for_scale)


x_transform_dt = scaler.transform(x_cols_for_scale)
x_scaled_df = pd.DataFrame(x_transform_dt, columns=col_for_scaled)
X.drop(col_for_scaled,axis=1,inplace=True)
X.reset_index(drop=True, inplace=True)
x_scaled_df.reset_index(drop=True, inplace=True)
X=pd.concat([X, x_scaled_df], axis=1)
X.head()

```

# Spliting into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2, random_state=50)
```

```python
from xgboost import XGBClassifier as xGBClassifier
from sklearn.metrics import accuracy_score
xgb = xGBClassifier()
xgb.fit(x_train,y_train)
xgbtd=xgb.predict(x_test)
xgbaccuracy = accuracy_score(y_test, xgbtd)
xgbaccuracy
```

```python
from sklearn.ensemble import RandomForestClassifier as randomForestClassifier
from sklearn.metrics import accuracy_score
rfc=randomForestClassifier()
rfc.fit(x_train,y_train)
y_pred=rfc.predict(x_test)
rfcaccuracy = accuracy_score(y_test, y_pred)
rfcaccuracy
```

```python
# make predictions on the testing set
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
y_pred = rfc.predict(x_test)

# calculate the evaluation metrics
print("accuracy",accuracy_score(y_test, y_pred))

#precision
print("precision",precision_score(y_test, y_pred))


#Recall
print("recall",recall_score(y_test, y_pred))


#f1 score
print("f1_score",f1_score(y_test, y_pred))


#confusion matrix
print("confusion matrix",confusion_matrix(y_test, y_pred))
```
